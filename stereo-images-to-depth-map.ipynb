```python
%matplotlib inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#Single-view Stereo Images 

#When imput stereo images are of different sizes
# width = 953
# height = 981
# dim = (width, height)
# imgL  = cv2.resize(cv2.imread('new-colour_l.jpg',0), dim, interpolation = cv2.INTER_AREA)
# imgR  = cv2.resize(cv2.imread('new-colour_r.jpg',0), dim, interpolation = cv2.INTER_AREA)

imgL = cv2.imread('color_l.png',0)
imgR = cv2.imread('color_r.png',0)
```

```python
#Calculating Depth Map from single-view Stereo Images
stereo = cv2.StereoBM_create(numDisparities=16, blockSize=15)
disparity = stereo.compute(imgL,imgR)
```

```python
#Display input images and output depth map
f = plt.figure(figsize=(22,8))
f.add_subplot(1,3, 1)
plt.title("Left stereo image")
plt.imshow(imgL,'gray', aspect='auto')

f.add_subplot(1,3, 2)
plt.title("Right stereo image")
plt.imshow(imgR,'gray', aspect='auto')

f.add_subplot(1,3, 3)
plt.title("Depth map response")
plt.imshow(disparity,'gray', aspect='auto')
plt.show()
```
